```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.interpolate import make_interp_spline
```

```python
# Create the DataFrame
to_plot_df = pd.DataFrame({
    "Model": ["GAE", "GAE", "GAE", "NB-VAE", "NB-VAE", "NB-VAE", "FlatVI", "FlatVI", "FlatVI"],
    "Latent dimension": [10., 20., 30., 10., 20., 30., 10., 20., 30.],
    "Velocity consistency": [0.81, 0.77, 0.75, 0.92, 0.89, 0.86, 0.94, 0.95, 0.94]
})

# Create a smooth plot
plt.figure(figsize=(4,3))
ax = sns.lineplot(data=to_plot_df, x="Latent dimension", y="Velocity consistency", hue="Model", palette="colorblind", linewidth = 3, 
            marker="o", markersize=7)
# ax = sns.lmplot(data=to_plot_df, x='Latent dimension', y='Velocity consistency', hue='Model', ci=None, order=2, truncate=True,
#           palette="colorblind",height=2.5, aspect=1.5)
plt.title("Velocity Consistency vs. Latent Dimension", fontsize=12)
plt.xlabel("Latent Dimension", fontsize=12)
plt.ylabel("Velocity Consistency", fontsize=12)
plt.ylim(0.70,1)
plt.grid(False)
plt.legend(title="Model", bbox_to_anchor=(1, 1))
# Show the plot
plt.savefig("consistency_plot.png", dpi=300, bbox_inches='tight')
```
